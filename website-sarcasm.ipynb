```python
!pip install gradio

import gradio as gr
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle

# Load your trained model (update the path if needed)
model = tf.keras.models.load_model("/kaggle/input/sarcasm_cnn_lstm.h5/pytorch/default/1/sarcasm_cnn_lstm.h5")

# Load your tokenizer (ensure you have saved it as a pickle file)
with open("/kaggle/input/tokenizer_sarcasm/other/default/1/tokenizer.pkl", "rb") as handle:
    tokenizer = pickle.load(handle)

MAX_SEQUENCE_LENGTH = 50  # Must match training

def predict_sarcasm(text):
    # Preprocess input text
    sequence = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(sequence, maxlen=MAX_SEQUENCE_LENGTH, padding='post')
    # Get prediction probability
    prediction = model.predict(padded)[0][0]
    return "Sarcastic" if prediction > 0.5 else "Not Sarcastic"

# Create a Gradio interface
iface = gr.Interface(fn=predict_sarcasm, inputs="text", outputs="text",
                     title="Sarcasm Detector",
                     description="Enter a sentence and the model will tell you if it's sarcastic.")

# Launch the interface with a public shareable link
iface.launch(share=True)
```
